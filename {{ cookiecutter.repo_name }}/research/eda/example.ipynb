```python
import pip
def import_or_install(package):
    try:
        __import__(package)
        print('{} is already installed'.format(package))
    except ImportError:
        pip.main(['install', package])  
```

```python
pkgs_list = ['numpy', 'pandas', 'matplotlib','seaborn']
for package in pkgs_list:
    import_or_install(package)
```

# Iris Dataset Exploration

1. Target Variable Distribution
2. Univarite Analysis
3. Multivariate analysis

```python
iris = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv')
display(iris.sample(5))
```

## Target Variable Distribution

```python
iris.species.value_counts()
```

## Univariate Analysis

```python
iris.hist(edgecolor='black', linewidth=1.5)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

## Multivariate Analysis

```python
# plotting constants
P_L = "petal_length"
P_W = "petal_width"
S_L = "sepal_length"
S_W = "sepal_width"
R = "red"
B = "blue"
G = "green"
```

```python
fig = iris[iris.species=='setosa'].plot.scatter(x = S_L, y = S_W, color = R, label = 'Setosa')
iris[iris.species=='versicolor'].plot.scatter(x = S_L, y = S_W,color = G, label = 'Versicolor',ax = fig)
iris[iris.species=='virginica'].plot.scatter(x = S_L, y = S_W, color = B, label = 'Virginica', ax = fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("-- Sepal Dimensions: --\ncolor-coded by Species")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
fig = iris[iris.species=='setosa'].plot.scatter(x = P_L,y = P_W, color = R, label='Setosa')
iris[iris.species=='versicolor'].plot.scatter(x = P_L, y = P_W, color = G, label='Versicolor',ax=fig)
iris[iris.species=='virginica'].plot.scatter(x = P_L, y = P_W, color = B, label='Virginica', ax=fig)
fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title("-- Petal Dimensions: --\ncolor-coded by Species")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```
